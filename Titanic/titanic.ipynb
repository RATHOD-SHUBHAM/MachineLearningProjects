## Name Feature

1. Extract the index of name
2. Replace the less know titles with a better known titles
3. convert the categorical data into numeric ones

```python
train_test_data = [train,test]
train_test_data
```

```python
for dataset in train_test_data:
    dataset['Title'] = dataset.Name.str.extract('([a-zA-Z]+)\.')
dataset
# https://pandas.pydata.org/docs/reference/api/pandas.Series.str.extract.html
```

```python
# now we replace some less common title with others
# un married men -- > master 
# married men --> Mr
# un married women -- > miss
# married women -- > mrs
```

```python
for dataset in train_test_data:
    dataset['Title'] = dataset['Title'].replace(['Capt','Countess','Col','Don','Dona','Dr','Jonkheer','Lady','Major','Rev','Sir'],'Others')
    dataset['Title'] = dataset['Title'].replace(['Mlle','Ms'],'Miss')
    dataset['Title'] = dataset['Title'].replace(['Mme'],'Mrs')
#     dataset['Title'] = dataset['Title'].replace(['Ms'],)
dataset
```

```python
for dataset in train_test_data:
    age_avg = dataset['Age'].mean()
    age_std = dataset['Age'].std()
    age_null_value = dataset['Age'].isnull().sum()
    
    age_null_to_random = np.random.randint(age_avg-age_std,age_avg+age_std,size = age_null_value)
    
    dataset['Age'][np.isnan(dataset['Age'])] = age_null_to_random
    dataset['Age'] = dataset['Age'].astype(int)
dataset
```
