```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
from sklearn.metrics import confusion_matrix
import numpy as np
from scipy.optimize import linear_sum_assignment
from sklearn.metrics import confusion_matrix
```

```python
def actual_domain(actual,predict):
    actual_d = list(set(actual))
    predict_d = list(set(predict))
    new_predict = []
    for x in predict:
        new_predict.append(actual_d[predict_d.index(x)])
    return new_predict
def _make_cost_m(cm):
    s = np.max(cm)
    return (s-cm)
```

```python
def km(qnumber, dataset, dataset_name, from_index, to_index,k):
    from sklearn.metrics import confusion_matrix as cm
    print("QUESTION "+str(qnumber))
   
    print(dataset_name)
    
    # Take all classes 0 to 10
    dataset_true = dataset.iloc[from_index:to_index, 0]
    
    # Taking all dataset for the corresponding classes 0 to 10
    dataset = dataset.iloc[from_index:to_index, 1:]
    dataset.shape
    
    # For 10 classes n_clusters = 10
   
    max_iter = 10
    min_obj_func=float('inf')
   
    for i in range(max_iter):
        init_kmean = KMeans(n_clusters=k, max_iter=1).fit(dataset)
        print('\nObj Func of iteration '+str(i+1))
        print(init_kmean.inertia_)
        if(init_kmean.inertia_<min_obj_func):
            min_obj_func = init_kmean.inertia_
            kmeans = init_kmean
   
    print("\n==================MIN OBJECTIVE FUNCTION============\n")
    print(min_obj_func)
   
    pred = actual_domain(dataset_true,kmeans.labels_)
    
    print('\n\npredictions')
    print(pred)
    
    print('\n\nconfusion matrix treated as bipartite graph')
    cm1 = pd.DataFrame(cm(dataset_true, pred))
    
    
    print("\n\nConfusion Matrix: ")
    print(cm1)
    
    print("\n\n")
    print('ACCURACY Before Bipartite Matching: ' + str(sum(np.diag(cm1)) / cm1.sum(numeric_only=True).sum()))
    print("\n\n")
    
    
    print('\nPerformaing Hungarian Algorithm for bipartite matching, as a result the confusion matrix is transformed')
    
    # Hungarian Algorithm for Bipartite matching
    indices = linear_sum_assignment(_make_cost_m(cm1))
    order_indices = indices[1]
    confusion_matrix = cm1.iloc[:, order_indices]
    confusion_matrix.columns = [i for i in range(len(confusion_matrix))]
    
    print("\n\nConfusion Matrix thats reordered: ")
    print(confusion_matrix)
    
    print("\n\n")
    print('ACCURACY: ' + str(sum(np.diag(confusion_matrix)) / confusion_matrix.sum(numeric_only=True).sum()))
    print("\n\n")

```

```python
handwritten = pd.read_csv("HandWrittenLetters.txt", header=None).T
km(3, handwritten, "Handwritten Dataset", 0, len(handwritten),26)
```
