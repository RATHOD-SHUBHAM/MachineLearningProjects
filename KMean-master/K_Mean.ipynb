```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean_1 = (5,5)
mean_2 = (-5,5)
mean_3 = (-5,-5)
co_varience_1 = [[4,0],[0,4]]
co_varience_2 = [[16,0],[0,16]]
```

```python
fig = plt.gcf()
x_1, y_1 = np.random.multivariate_normal(mean_1, co_varience_1, 100).T
plt.plot(x_1, y_1, 'bo',alpha=0.5)
x_2, y_2 = np.random.multivariate_normal(mean_2, co_varience_1, 100).T
plt.plot(x_2, y_2,'ro',alpha=0.5)
x_3, y_3 = np.random.multivariate_normal(mean_3, co_varience_1, 100).T
plt.plot(x_3, y_3, 'go',alpha=0.5)
plt.axis('equal')
plt.show()
fig.savefig('input.png',dpi = 200)
```

```python
fig = plt.gcf()
x_4, y_4 = np.random.multivariate_normal(mean_1, co_varience_2, 100).T
plt.plot(x_4, y_4, 'bo',alpha=0.5)
x_5, y_5 = np.random.multivariate_normal(mean_2, co_varience_2, 100).T
plt.plot(x_5, y_5,'ro',alpha=0.5)
x_6, y_6 = np.random.multivariate_normal(mean_3, co_varience_2, 100).T
plt.plot(x_6, y_6, 'go',alpha=0.5)
plt.axis('equal')
plt.show()
fig.savefig('input2.png',dpi = 200)
```

```python
input_x1 = np.concatenate((x_1,x_2,x_3))
```

```python
input_y1 = np.concatenate((y_1,y_2,y_3))
```

```python
input_x2 = np.concatenate((x_4,x_5,x_6))
```

```python
input_y2 = np.concatenate((y_4,y_5,y_6))
```

```python
import pandas as pd
train_set_1 = pd.DataFrame([input_x1, input_y1]).T
```

```python
import pandas as pd
train_set_2 = pd.DataFrame([input_x2, input_y2]).T
```

```python
x1_c0 = train_set_1.iloc[:,0]
y1_c1 = train_set_1.iloc[:,1]
```

```python
x2_c0 = train_set_2.iloc[:,0]
y2_c1 = train_set_2.iloc[:,1]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3).fit(train_set_1)
prediction_1 = kmeans.labels_
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3).fit(train_set_2)
prediction_2 = kmeans.labels_
```

```python
fig = plt.gcf()
plt.scatter(np.array(x1_c0), np.array(y1_c1), alpha=0.5, c=prediction_1, cmap='rainbow')
fig.savefig('output1.png',dpi = 200)
```

```python
fig = plt.gcf()
plt.scatter(np.array(x2_c0), np.array(y2_c1), alpha=0.5, c=prediction_2, cmap='rainbow')
fig.savefig('output.png',dpi = 200)
```
